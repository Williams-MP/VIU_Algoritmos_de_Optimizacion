# Clases de Complejidad: P vs NP

En ciencias de la computación, clasificamos los problemas según qué tan "difíciles" son para un ordenador.

---

### 1. Clase P (Polinómicos)
Son los problemas que un ordenador puede **resolver** de forma eficiente en un tiempo razonable.
* **Definición:** El tiempo de ejecución crece según un polinomio ($n, n^2, n^3$, etc.).
* **Ejemplo:** Buscar un elemento en una lista, ordenar nombres alfabéticamente o el código de tu ejemplo ($O(n^2)$).
* **Veredicto:** Son problemas **tratables** (fáciles de manejar).

### 2. Clase NP (Polinómico No determinista)
Son problemas que son muy difíciles de resolver, pero muy **fáciles de verificar**.
* **Definición:** Si alguien te da una "solución candidata", puedes comprobar si es correcta en un tiempo eficiente (P). Sin embargo, encontrar esa solución puede llevar eones.
* **Ejemplo:** Resolver un Sudoku, encontrar los factores primos de un número gigante (base de la criptografía) o el "Problema del Viajante".
* **Veredicto:** Son problemas que hoy consideramos **intrínsecamente difíciles**.

---

### El dilema: ¿P = NP?
Es el problema más importante de la informática actual.
* **Si P ≠ NP (Creencia mayoritaria):** Existen problemas cuya solución es fácil de comprobar pero imposible de encontrar rápidamente. La seguridad en internet se basa en esta premisa.
* **Si P = NP:** Significaría que existe un "atajo" para todos los problemas NP que aún no hemos descubierto. Si se demostrara, ninguna contraseña actual sería segura.

# Algoritmos de Recorrido de Grafos: BFS vs DFS

Existen dos estrategias principales para explorar nodos en un grafo o árbol. La elección depende de si queremos buscar "a lo ancho" o "a lo profundo".

---

## 1. BFS (Breadth-First Search) - Búsqueda en Amplitud
El BFS explora el grafo como una **onda expansiva**. Visita primero todos los nodos vecinos directos antes de pasar al siguiente nivel de distancia.

* **Estrategia:** Horizontal (Nivel por nivel).
* **Estructura de datos:** Usa una **Cola (Queue)** (Sistema FIFO: el primero en entrar es el primero en salir).
* **Complejidad:** $O(V + E)$, donde $V$ es vértices y $E$ aristas.
* **Uso ideal:** * Encontrar la **ruta más corta** (en grafos sin pesos).
    * Redes sociales (amigos de 1er, 2º y 3er grado).
    * Sistemas de navegación GPS.

---

## 2. DFS (Depth-First Search) - Búsqueda en Profundidad
El DFS actúa como un explorador en un laberinto. Elige un camino y llega **hasta el fondo** antes de retroceder (backtracking) para intentar otra ruta.

* **Estrategia:** Vertical (Hacia lo profundo).
* **Estructura de datos:** Usa una **Pila (Stack)** o **Recursividad** (Sistema LIFO: el último en entrar es el primero en salir).
* **Complejidad:** $O(V + E)$.
* **Uso ideal:**
    * Detección de **ciclos** en un grafo.
    * Resolución de puzzles o laberintos (probar todas las opciones).
    * Ordenamiento topológico (programación de tareas dependientes).

---

## 3. Tabla Comparativa

| Característica | BFS (Amplitud) | DFS (Profundidad) |
| :--- | :--- | :--- |
| **Movimiento** | Por niveles (vecinos primero) | Por ramas (hijos primero) |
| **Estructura** | Cola (`Queue`) | Pila (`Stack`) o Recursión |
| **Memoria** | Consume más en grafos muy anchos | Consume más en grafos muy profundos |
| **Objetivo** | Encontrar lo más cercano / corto | Explorar todas las posibilidades |

---

## Ejemplo Visual
Si tenemos un nodo raíz **A** conectado a **B** y **C**, y estos a su vez tienen hijos:

1.  **BFS visitaría:** A → [B, C] → [Hijos de B] → [Hijos de C].
2.  **DFS visitaría:** A → B → [Hijo de B] → [Nieto de B] ... y luego vuelve para ir a C.
